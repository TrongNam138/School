```python
# lấy các phần tử trong mảng theo yêu cầu

import numpy as np

row = 6
col = 8
 

m = np.random.randint(0, 100, [row, col])

v_m = []
v_s = []

print(m, '\n')
print('Kiểu dữ liệu: ', m.dtype)
print('Kích thước: ', m.shape)
print('Số phần tử: ', m.size)
print('Số chiều: ', m.ndim)

for i in range(0, min(row, col)):
    v_m.append(m[i, i])
    v_s.append(m[min(row, col) - i - 1, i])

v_m = np.array(v_m)
v_s = np.array(v_s)


print('Các phần tử nằm trền đường chéo chính: \n', v_m)
print('Các phần tử nằm trền đường chéo phụ: \n', v_s)

x = int(input('\nNhập vào giá trị x (0-100): '))

s_x = 0
e_x = 0
l_x = 0

for r in range(0, row):
    for c in range(0, col):
        if m[r, c] < x: 
            s_x += 1;
        elif m[r, c] == x:
            e_x += 1;
        else:
            l_x += 1;
            
print('----------------------')
print('Số phần tử < x: ', s_x)
print('Số phần tử = x: ', e_x)
print('Số phần tử > x: ', l_x)
```

```python
# phân tích dữ liệu điểm
import numpy as np

scores = np.loadtxt('./data/scores.txt', delimiter = ',', dtype = np.int8)
print(scores, '\n')

scores_mean_std = np.array([])
student_std_max = [0, scores[:, 0].std()]
student_std_min = [0, scores[:, 0].std()]
for i in range(0, scores.shape[1]):
    scores_mean_std = np.append(scores_mean_std, scores[:, i].mean())
    
    i_std = scores[:, i].std()
    if(i_std > student_std_max[1]):
        student_std_max[0] = i
        student_std_max[1] = i_std
    
    if(i_std < student_std_min[1]):
        student_std_min[0] = i
        student_std_min[1] = i_std

scores_mean_sub = np.array([])
subject_std_max = [0, scores[0, :].std()]
subject_std_min = [0, scores[0, :].std()]
for i in range(0, scores.shape[0]):
    scores_mean_sub = np.append(scores_mean_sub, scores[i, :].mean())
    
    i_std = scores[i, :].std()
    if(i_std > subject_std_max[1]):
        subject_std_max[0] = i
        subject_std_max[1] = i_std
    
    if(i_std < subject_std_min[1]):
        subject_std_min[0] = i
        subject_std_min[1] = i_std
    
print('Điểm trung bình của từng học sinh trong lớp: \n', scores_mean_std)
print('------------------------------------------------------------')
print('Điểm trung bình cao nhất: ', scores_mean_std.max())
print('Của học sinh thứ: ', scores_mean_std.argmax() + 1)
print('Bảng điểm đầy đủ của học sinh: \n', scores[:, scores_mean_std.argmax()])
print('------------------------------------------------------------')
print('Điểm trung bình thấp nhất: ', scores_mean_std.min())
print('Của học sinh thứ: ', scores_mean_std.argmin() + 1)
print('Bảng điểm đầy đủ của học sinh: \n', scores[:, scores_mean_std.argmin()])

print('\n============================================================\n')

print('Điểm trung bình của từng môn học: \n', scores_mean_sub)
print('------------------------------------------------------------')
print('Điểm trung bình cao nhất: ', scores_mean_sub.max())
print('Của môn học thứ: ', scores_mean_sub.argmax() + 1)
print('Bảng điểm đầy đủ của môn học: \n', scores[scores_mean_sub.argmax(), :])
print('------------------------------------------------------------')
print('Điểm trung bình thấp nhất: ', scores_mean_sub.min())
print('Của môn học thứ: ', scores_mean_sub.argmin() + 1)
print('Bảng điểm đầy đủ của môn học: \n', scores[scores_mean_sub.argmin(), :]) 

print('\n============================================================\n')
print('Học sinh học đồng đều nhất: ', student_std_min[0])
print('Độ lệch: ', student_std_min[1])
print('Bảng điểm của học sinh học đồng đều: ', scores[:, student_std_min[0]])
print('------------------------------------------------------------')
print('Học sinh học lệch nhất: ', student_std_max[0])
print('Độ lệch: ', student_std_max[1])
print('Bảng điểm của học sinh học lệch: ', scores[:, student_std_max[0]])

print('\n')

print('Môn học có điểm đồng đều nhất: ', subject_std_min[0])
print('Độ lệch: ', subject_std_min[1])
print('Bảng điểm của môn học đồng đều: ', scores[subject_std_min[0], :])
print('------------------------------------------------------------')
print('Học sinh học lệch nhất: ', subject_std_max[0])
print('Độ lệch: ', subject_std_max[1])
print('Bảng điểm của Môn học lệch: ', scores[subject_std_max[0], :])
```
