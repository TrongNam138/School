```python
import pandas as pd
import matplotlib.pyplot as plt

gas_prices = pd.read_csv("./data/gas_prices.csv")

plt.figure(figsize=(16, 4))

plt.plot(
    list(gas_prices['Year']),
    list(gas_prices['USA']),
)

plt.grid(ls='--')

plt.title('Gi gas ca nc m')
plt.xticks(list(gas_prices['Year']), rotation=0)
plt.show()
```
